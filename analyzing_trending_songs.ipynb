```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'.\predicted_data\predictions.csv')
```

```python
print("Summary Statistics:")
print(df['predicted_mood'].describe())

# Count of each mood
print("\nMood Frequency:")
print(df['predicted_mood'].value_counts())

# Histogram of moods
plt.figure(figsize=(8, 6))
plt.hist(df['predicted_mood'], bins=range(0, 11), edgecolor='black', alpha=0.7)
plt.title('Distribution of Predicted Moods')
plt.xlabel('Predicted Mood')
plt.ylabel('Frequency')
plt.xticks(range(0, 11))
plt.show()

# Bar chart of mood frequency
plt.figure(figsize=(8, 6))
df['predicted_mood'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Mood Frequency by Value')
plt.xlabel('Predicted Mood')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

```
